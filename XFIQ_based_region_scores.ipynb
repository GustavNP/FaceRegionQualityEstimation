```python
import pandas as pd
import numpy as np
import statistics as stat

import cv2
import os.path
```

```python
# filePath_mask = "C:\\Users\\admin\\source\\repos\\OFIQ-Project\\install_x86_64\\Release\\bin\\face_parsing_images\\face_parsing_mask.png"
# filePath_mask = "C:\\Users\\admin\\source\\repos\\OFIQ-Project\\install_x86_64\\Release\\bin\\face_parsing_images\\face_parsing_cap.png"
filePath_mask = "C:\\Users\\admin\\source\\repos\\OFIQ-Project\\install_x86_64\\Release\\bin\\face_parsing_images\\face_parsing_c-01-frontal.png"

region_mask_img = cv2.imread(filePath_mask)
region_mask_img = cv2.cvtColor(region_mask_img, cv2.COLOR_BGR2GRAY)

cv2.imwrite('test_region_christoph.png', region_mask_img)
```

```python
# file_path_pixel_level = "C:\\Users\\admin\\source\\repos\\ExplainableFaceImageQuality\\src\\plots\\upscaled-grad-mask.png"
# file_path_pixel_level = "C:\\Users\\admin\\source\\repos\\ExplainableFaceImageQuality\\src\\plots\\upscaled-grad-cap.png"
file_path_pixel_level = "C:\\Users\\admin\\source\\repos\\ExplainableFaceImageQuality\\src\\plots\\upscaled-grad-aligned_c-01-frontal.png"

img_pixel_level = cv2.imread(file_path_pixel_level)
img_pixel_level = cv2.cvtColor(img_pixel_level, cv2.COLOR_BGR2GRAY)

cv2.imwrite('test_pixel_level_christoph.png', img_pixel_level)
```

```python
pixel_list = []

for i in range(0,256):
    pixel_list.append([])


for i in range(0,400):
    for k in range(0,400):
        region_class_number = region_mask_img[i, k]
        pixel_quality_value = img_pixel_level[i, k]
        pixel_list[region_class_number].append(pixel_quality_value)

print(pixel_list)
```

```python

class_quality_means = []
for i in range(0,256):
    if len(pixel_list[i]) == 0:
        class_quality_means.append(255) # 255 is the worst value with the current settings
    else:
        class_quality_means.append(int(stat.fmean(pixel_list[i])))

print(class_quality_means)
```

```python
# create copy of class region mask
# for every pixel, check class and access mean value in the mean value list and assign that as the new pixel value

averaged_regions_image = region_mask_img.copy()

for i in range(0,400):
    for k in range(0,400):
        averaged_regions_image[i,k] = class_quality_means[region_mask_img[i,k]]

cv2.imwrite('averaged_regions_christoph.png', averaged_regions_image)
```

```python
output_heatmap = cv2.applyColorMap(averaged_regions_image, cv2.COLORMAP_RAINBOW)
cv2.imwrite('regions_heatmap_christoph.png', output_heatmap)
```
